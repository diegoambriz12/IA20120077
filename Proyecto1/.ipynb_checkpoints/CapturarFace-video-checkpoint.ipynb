```python
import cv2 as cv
import numpy as np
import os

# Definir las emociones y asignar etiquetas numéricas
emociones = {'triste': 0, 'feliz': 1, 'sorprendido': 2}

dataSet = 'C:/Users/diego/OneDrive/Documentos/IA/Proyecto1/caras'
labels = []
facesData = []

for emocion, etiqueta in emociones.items():
    subdirectoryPath = os.path.join(dataSet, emocion)
    for faceName in os.listdir(subdirectoryPath):
        imagePath = os.path.join(subdirectoryPath, faceName)
        image = cv.imread(imagePath, 0)
        if image is not None:
            labels.append(etiqueta)
            facesData.append(image)

print("Número de imágenes por emoción:")
for emocion, etiqueta in emociones.items():
    print(f"{emocion}: {labels.count(etiqueta)}")

# Verificar que tenemos datos suficientes para cada emoción
assert all(labels.count(etiqueta) > 0 for etiqueta in emociones.values()), "No hay suficientes datos para algunas emociones"

# Crear el modelo y entrenarlo
faceRecognizer = cv.face.EigenFaceRecognizer_create()
faceRecognizer.train(facesData, np.array(labels))
faceRecognizer.write('emocionesface.xml')
```
